```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#array x goes from 0 to 1 in 100 steps
x = np.linspace(0,1,100)
#function that returns exp(x)
def retexp(y):
    return np.exp(y)
```

```python
#y now equals the exp of x.
y = retexp(x)
```

```python
#plot x vs. y
#x = "Time[milliseconds]"
#y = "Awesomeness"
#save figure as PDF.

plt.plot(x,y)
plt.xlabel("Time[milliseconds]")
plt.ylabel("Awesomeness")
plt.savefig('journal-two-prompt-one.png', bbox_inches="tight",dpi=600)
```
